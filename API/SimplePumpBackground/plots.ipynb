```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
# strip "19 ml water" from Sample column
data['Sample'] = data['Sample'].str.replace('19 ml water + ', '').str.strip()
data['Sample'] = data['Sample'].str.replace('19 ml base water + ', '').str.strip()
data
```

```python
# FacetGrid: RGB values by Sample and Lighting
import seaborn as sns
import matplotlib.pyplot as plt

# Melt the dataframe for easier plotting
melted = data.melt(id_vars=['Sample', 'Lighting'], value_vars=['R', 'G', 'B'], var_name='Channel', value_name='Intensity')

# Create a FacetGrid: one row per Lighting, one column per Channel
g = sns.FacetGrid(melted, row='Lighting', col='Channel', hue='Sample', margin_titles=True, height=4, aspect=1.2)
g.map(sns.stripplot, 'Sample', 'Intensity', dodge=True, jitter=True, alpha=0.8)
g.add_legend(title='Sample')
g.set_xticklabels(rotation=30, ha='right')
g.set_axis_labels('Sample', 'Intensity')
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle('RGB Intensity by Sample and Lighting', fontsize=16)
plt.show()
```

```python
# Create a 2x2 grid of scatter plots for RGB channels by Sample and Lighting
samples = data['Sample'].unique()
rgb_channels = ['R', 'G', 'B']
light = data['Lighting'].unique()

rows, cols = 2, 2
fig, axes = plt.subplots(rows, cols, figsize=(14, 10))
axes = axes.flatten()

for i, channel in enumerate(light):
    ax = axes[i]
    for color, rgb in zip(['red', 'green', 'blue'], rgb_channels):
        for sample in samples:
            sample_data = data[(data['Lighting'] == channel) & (data['Sample'] == sample)]
            ax.scatter([sample]*len(sample_data), sample_data[rgb], color=color, label=f'{rgb} - {sample}' if i == 0 else "", s=60, alpha=0.7, edgecolor='k')
    ax.set_title(f'Scatter of RGB by Sample under {channel} Lighting')
    ax.set_xlabel('Sample')
    ax.set_ylabel('Intensity')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=30, ha='right')

handles, labels = axes[0].get_legend_handles_labels()

# Hide unused subplot if less than 4 channels
for j in range(len(light), rows * cols):
    fig.delaxes(axes[j])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, y=data.R, x=data.columns[1], hue='Sample')
plt.title('Data by Sample Type')
plt.xlabel('Light')
plt.ylabel('R Value')
plt.legend(title='Sample Type')
plt.show()
```

```python
# Grouped bar plot: RGB values by Sample and Lighting
import matplotlib.pyplot as plt
import numpy as np

# Prepare data for grouped bar plot
samples = data['Sample'].unique()
lightings = data['Lighting'].unique()
rgb_channels = ['R', 'G', 'B']

bar_width = 0.2
x = np.arange(len(samples))

fig, axes = plt.subplots(1, len(lightings), figsize=(6 * len(lightings), 6), sharey=True)
if len(lightings) == 1:
    axes = [axes]
for idx, lighting in enumerate(lightings):
    ax = axes[idx]
    lighting_data = data[data['Lighting'] == lighting]
    for i, channel in enumerate(rgb_channels):
        values = [lighting_data[lighting_data['Sample'] == sample][channel].mean() for sample in samples]
        ax.bar(x + i * bar_width, values, width=bar_width, label=channel, color=['red', 'green', 'blue'][i], alpha=0.7)
    ax.set_xticks(x + bar_width)
    ax.set_xticklabels(samples, rotation=30, ha='right')
    ax.set_title(f'Lighting: {lighting}')
    ax.set_xlabel('Sample')
    ax.set_ylabel('Mean Intensity')
    ax.legend(title='Channel')
plt.tight_layout()
plt.show()
```

```python
# Grouped bar plot with error bars (variance) for RGB values by Sample and Lighting
import matplotlib.pyplot as plt
import numpy as np

samples = data['Sample'].unique()
lightings = data['Lighting'].unique()
rgb_channels = ['R', 'G', 'B']

bar_width = 0.2
x = np.arange(len(samples))

fig, axes = plt.subplots(1, len(lightings), figsize=(6 * len(lightings), 6), sharey=True)
if len(lightings) == 1:
    axes = [axes]
for idx, lighting in enumerate(lightings):
    ax = axes[idx]
    lighting_data = data[data['Lighting'] == lighting]
    for i, channel in enumerate(rgb_channels):
        means = [lighting_data[lighting_data['Sample'] == sample][channel].mean() for sample in samples]
        stds = [lighting_data[lighting_data['Sample'] == sample][channel].std() for sample in samples]
        ax.bar(x + i * bar_width, means, width=bar_width, label=channel, color=['red', 'green', 'blue'][i], alpha=0.7, yerr=stds, capsize=6)
    ax.set_xticks(x + bar_width)
    ax.set_xticklabels(samples, rotation=30, ha='right')
    ax.set_title(f'{lighting}')
    ax.set_xlabel('Sample')
    ax.set_ylabel('Mean Intensity  Std')
    ax.legend(title='Channel')
plt.tight_layout()
plt.show()
```
